```python
# Rock, Paper, Scissors â€” Score Tracking + Best-of + Emoji Version
import random

choices = ["rock", "paper", "scissors"]
short_map = {"r": "rock", "p": "paper", "s": "scissors"}
emoji_map = {"rock": "ðŸª¨", "paper": "ðŸ“„", "scissors": "âœ‚ï¸"}

def get_player_choice():
    while True:
        raw = input("Enter rock/paper/scissors (or r/p/s). Type 'q' to quit: ").strip().lower()
        if raw in ("q", "quit", "exit"):
            return None
        if raw in short_map:
            return short_map[raw]
        if raw in choices:
            return raw
        print("Invalid input â€” please enter rock, paper, scissors, r, p, s, or q to quit.")

def play_one_round(show_emoji=False):
    player = get_player_choice()
    if player is None:               # user chose to quit
        return None, None, None
    computer = random.choice(choices)
    result = "tie"
    if player == computer:
        result = "tie"
    elif (player == "rock" and computer == "scissors") or \
         (player == "scissors" and computer == "paper") or \
         (player == "paper" and computer == "rock"):
        result = "win"
    else:
        result = "lose"

    # display choices
    if show_emoji:
        print(f"\nYou: {player} {emoji_map[player]}")
        print(f"CPU: {computer} {emoji_map[computer]}")
    else:
        print(f"\nYou: {player}")
        print(f"CPU: {computer}")

    # display result
    if result == "tie":
        print("Result: It's a tie! ðŸ˜\n")
    elif result == "win":
        print("Result: You win! ðŸŽ‰\n")
    else:
        print("Result: You lose. ðŸ˜¢\n")

    return result, player, computer

def unlimited_match(show_emoji=False):
    player_score = 0
    cpu_score = 0
    round_no = 0
    print("\nStarting unlimited match. Type 'q' to quit anytime.\n")
    while True:
        round_no += 1
        print(f"--- Round {round_no} ---")
        result, _, _ = play_one_round(show_emoji=show_emoji)
        if result is None:  # player quit
            print("\nMatch ended by player.")
            break
        if result == "win":
            player_score += 1
        elif result == "lose":
            cpu_score += 1
        print(f"Score -> You: {player_score} | Computer: {cpu_score}\n")
    print(f"Final Score -> You: {player_score} | Computer: {cpu_score}")

def best_of_n_match(n, show_emoji=False):
    if n <= 0 or n % 2 == 0:
        print("Best-of game requires a positive odd number (e.g., 3, 5, 7).")
        return
    needed = n // 2 + 1
    player_score = 0
    cpu_score = 0
    round_no = 0
    print(f"\nStarting best-of-{n}. First to {needed} wins. Type 'q' to quit anytime.\n")
    while player_score < needed and cpu_score < needed:
        round_no += 1
        print(f"--- Round {round_no} ---")
        result, _, _ = play_one_round(show_emoji=show_emoji)
        if result is None:
            print("\nMatch aborted by player.")
            return
        if result == "win":
            player_score += 1
        elif result == "lose":
            cpu_score += 1
        print(f"Score -> You: {player_score} | Computer: {cpu_score}\n")

    if player_score > cpu_score:
        print(f"Match result: You win the best-of-{n}! ðŸ†")
    else:
        print(f"Match result: Computer wins the best-of-{n}. Better luck next time!")

def main():
    print("Welcome to Rock, Paper, Scissors!")
    print("Modes:")
    print("  1 - Unlimited rounds (play until you quit)")
    print("  2 - Best-of-N (enter an odd number, e.g., 3, 5, 7)")
    print("Emoji option: show emojis for choices (rock/paper/scissors)\n")

    # choose mode
    while True:
        mode = input("Choose mode (1 or 2). Type 'q' to quit: ").strip().lower()
        if mode in ("q", "quit", "exit"):
            print("Goodbye!")
            return
        if mode not in ("1", "2"):
            print("Please enter 1 or 2 (or q to quit).")
            continue

        # emoji preference
        emoji_pref = input("Show emojis? (y/n): ").strip().lower()
        show_emoji = emoji_pref in ("y", "yes")

        if mode == "1":
            unlimited_match(show_emoji=show_emoji)
            break
        else:
            n_raw = input("Enter an odd number for best-of (e.g., 3, 5, 7): ").strip()
            if n_raw in ("q", "quit", "exit"):
                print("Goodbye!")
                return
            if not n_raw.isdigit():
                print("That's not a positive integer. Returning to mode selection.")
                continue
            n = int(n_raw)
            best_of_n_match(n, show_emoji=show_emoji)
            break

if __name__ == "__main__":
    main()
```
